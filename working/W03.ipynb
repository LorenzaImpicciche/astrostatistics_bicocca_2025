```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
x_values = np.array([0, 1, 2, 3, 4])
frequencies = np.array([109, 65, 22, 3, 1])
N = sum(frequencies) 
```

```python
# Estimate lambda (mean)
lambda_estimate = np.sum(x_values * frequencies) / N

# Poisson distribution with estimated lambda
x_range = np.arange(0, 5)
poisson_probs = poisson.pmf(x_range, lambda_estimate) * N  # Normalize to 200 groups
```

```python
# Plot
plt.bar(x_values, frequencies, alpha=0.6, label='Data')
plt.plot(x_range, poisson_probs, label=f'Poisson (Î»={lambda_estimate:.2f})')

plt.xlabel('Deaths per group')
plt.ylabel('Number of groups')
plt.title('Horse-kick Deaths Distribution')
plt.legend()
```
